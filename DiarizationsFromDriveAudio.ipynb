```python
def getDiarizations(directory=None):
  pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization@2.1",
                                    use_auth_token="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")
  diarizations = []

  # List of audio file extensions
  audio_extensions = ['*.mp3', '*.wav', '*.flac', '*.aac', '*.ogg', '*.m4a', '*.wma']
  i = 0
  # Function to extract numerical values from filenames for sorting
  def numerical_sort_key(filename):
      # Extract numeric values from the filename using regex, or use the filename itself
      return [int(text) if text.isdigit() else text.lower() for text in re.split(r'(\d+)', filename)]

  # Get all files in the directory and sort them numerically
  files_in_directory = sorted(os.listdir(directory), key=numerical_sort_key)
  for filename in files_in_directory:
      for ext in audio_extensions:
            if fnmatch.fnmatch(filename, ext):
                # try:
                  print(filename)
                  start_end = []
                  start_end.append(filename)
                  diarization = pipeline(os.path.join(directory, filename), num_speakers=1)

                  diarization_full  = diarization.support(collar=0.8)

                  for turn, _, speaker in diarization_full.itertracks(yield_label=True):
                      start_end.append((turn.start, turn.end))
                  print(start_end)
                  diarizations.append(start_end)
                # except:
                #   print("Error you silly")


  return diarizations
```
